### STEP-4 = Feature Engineering

```python
## Train_Test_Split
x=df.drop('CO2 Emissions(g/km)',axis=1)
y=df['CO2 Emissions(g/km)']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=5,test_size=0.30)
```

```python
x_train.shape
```

### Step-Model training

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

### Step - Model Evaluation

```python
# Evaluation on testing data
y_test_pred=model.predict(x_test)
def test_accuracy(actual,predicted):
    Test_accuracy_score=r2_score(actual,predicted)
    print(f'Testing accuracy score = {Test_accuracy_score}')
    Test_MSA=mean_squared_error(actual,predicted)
    print(f'Testing mean squared error = {Test_MSA}')
    Test_MAE=mean_absolute_error(actual,predicted)
    print(f'Testing mean absolute error = {Test_MAE}')
    Test_RMSE=np.sqrt(Test_MSA)
    print(f'Testing root mran squared erroe = {Test_RMSE}')
test_accuracy(y_test,y_test_pred)
```

```python
# Evaluation on Training data
y_train_pred=model.predict(x_train)
def train_accuracy(actual,predicted):
    Train_accuracy_score=r2_score(actual,predicted)
    print(f'Training accuracy score = {Train_accuracy_score}')
    Train_MSA=mean_squared_error(actual,predicted)
    print(f'Training mean squared error = {Train_MSA}')
    Train_MAE=mean_absolute_error(actual,predicted)
    print(f'Training mean absolute error = {Train_MAE}')
    Train_RMSE=np.sqrt(Train_MSA)
    print(f'Testing root mran squared erroe = {Train_RMSE}')
train_accuracy(y_train,y_train_pred)
```

### Normality Test

```python
Residual_train = y_train - y_train_pred
Residual_train
```

```python
Residual_test = y_test - y_test_pred
Residual_test
```
